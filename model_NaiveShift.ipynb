```python
import pandas as pd
import numpy as np
import feature_functions as ff
from little_helpers import *

from sklearn.metrics import mean_absolute_error
```

```python
#import train, validation and test data
train_df, validation_df, test_df = pd.read_csv('data/train_fil_3.csv',index_col=0),pd.read_csv('data/validation_fil_3.csv',index_col=0),pd.read_csv('data/test_fil_3.csv',index_col=0)
for df in [train_df,validation_df,test_df]:
    df.columns = [int(col) for col in df.columns]
    df.index = pd.to_datetime(df.index)
```

```python
def naive_base_shift(y_true,shift=7):
    y_predict = y_true.shift(shift)
    return y_predict
```

```python
for i in validation_df.columns:
    shift = 7
    y_predict = naive_base_shift(validation_df[i],shift=shift)
    y_true = validation_df[i]
    #plot_prediction(y_true,y_predict,title = f'Article {i}')
    #plot_residuals(residuals(y_true,y_predict))
    #plot_residuals(pct_residuals(y_true,y_predict))
    print(f'Naive Model Performance for Article {i}: \n MAPE = {mape(y_true[shift:],y_predict[shift:])}\n MAE = {mean_absolute_error(y_true[shift:],y_predict[shift:])}')
```
